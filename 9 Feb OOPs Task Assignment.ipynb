```python
Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

Sure, here's how you can create a `Vehicle` class in Python with an `__init__` method that initializes instance variables for the name of the vehicle, maximum speed, and average speed:

class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_speed):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_speed = average_speed

# Example usage
car = Vehicle("Car", 200, 100)
print("Name of vehicle:", car.name_of_vehicle)
print("Max speed:", car.max_speed)
print("Average speed:", car.average_speed)
```

```python
Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

Certainly! Below is the implementation of the `Car` class, which is a child class of the `Vehicle` class. It inherits the attributes `name_of_vehicle`, `max_speed`, and `average_speed` from the `Vehicle` class and adds a method `seating_capacity`:

class Car(Vehicle):
    def seating_capacity(self, capacity):
        return f"The {self.name_of_vehicle} has a seating capacity of {capacity}."

# Example usage
car = Car("SUV", 180, 80)
print(car.seating_capacity(5))
```

```python
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance is a feature in object-oriented programming languages that allows a class to inherit attributes and methods from more than one parent class. This means that a subclass can inherit from multiple superclasses, thereby inheriting their properties and behaviors.

In Python, multiple inheritance is supported, allowing a class to inherit from multiple parent classes. When a class inherits from multiple parent classes, it inherits attributes and methods from all of them.

Here's a Python code example to demonstrate multiple inheritance:

# Parent class 1
class Animal:
    def speak(self):
        return "Animal speaks"

# Parent class 2
class Bird:
    def fly(self):
        return "Bird flies"

# Child class inheriting from both Animal and Bird
class Parrot(Animal, Bird):
    def sing(self):
        return "Parrot sings"

# Create an instance of Parrot class
parrot = Parrot()

# Demonstrate multiple inheritance
print(parrot.speak())  # Output: Animal speaks
print(parrot.fly())    # Output: Bird flies
print(parrot.sing())   # Output: Parrot sings
```

```python
Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

In Python, getters and setters are methods used to access and modify the attributes of a class. They provide controlled access to the class attributes, allowing validation and manipulation of data before setting or getting their values. This encapsulation helps maintain the integrity of the object's data and enables better control over attribute access.

A getter method is used to retrieve the value of an attribute, while a setter method is used to modify or set the value of an attribute. In Python, we typically use property decorators to define getter and setter methods for attributes.

Here's how you can create a class with a getter and a setter method:

class MyClass:
    def __init__(self):
        self._my_attribute = None  # Initialize attribute to None

    # Getter method
    @property
    def my_attribute(self):
        return self._my_attribute

    # Setter method
    @my_attribute.setter
    def my_attribute(self, value):
        if isinstance(value, int):  # Perform validation
            self._my_attribute = value
        else:
            raise ValueError("Attribute must be an integer.")

# Example usage
obj = MyClass()

# Set attribute value using the setter method
obj.my_attribute = 42

# Get attribute value using the getter method
print(obj.my_attribute)  # Output: 42

# Try to set attribute value with a non-integer value (will raise an error)
# obj.my_attribute = "Invalid"  # Uncomment this line to see the ValueError
```

```python
Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Method overriding in Python occurs when a subclass provides a specific implementation of a method that is already defined in its superclass. This allows the subclass to customize or extend the behavior of the method inherited from the superclass. Method overriding is a form of polymorphism in object-oriented programming.

Here's a Python code example to demonstrate method overriding:

# Parent class
class Animal:
    def make_sound(self):
        return "Animal makes a generic sound"

# Child class overriding the make_sound method
class Dog(Animal):
    def make_sound(self):
        return "Dog barks"

# Child class overriding the make_sound method
class Cat(Animal):
    def make_sound(self):
        return "Cat meows"

# Create instances of Dog and Cat classes
dog = Dog()
cat = Cat()

# Demonstrate method overriding
print(dog.make_sound())  # Output: Dog barks
print(cat.make_sound())  # Output: Cat meows
```
